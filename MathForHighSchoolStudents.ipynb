```python
#Differential Equations
f = symbols('f',cls=Function) #重要事前工作
f(x)
```

```python
f(x).diff(x)
```

```python
f(x).diff(x,2)
```

```python
a=symbols('a') #可以顯示a(變數而不用計算出來)
diffeq = Eq(f(x)-x*f(x).diff(x)-a*(f(x)**2+f(x).diff(x)),0) #Eq....=0
diffeq
```

```python
dsolve(diffeq,f(x)) #解
```

```python
diffeq = Eq(f(x).diff(x,2)+f(x).diff(x)-6*f(x),0)
diffeq
```

```python
dsolve(diffeq,f(x)) #dsolve=解
```

```python
diffeq = Eq(f(x).diff(x,2)+5*f(x).diff(x)+6*f(x),3*exp(-2*x))
diffeq
```

```python
dsolve(diffeq)
```
